```python
# Load the results from both VADER and TextBlob analyses
today_date = datetime.today().strftime('%Y-%m-%d')
vader_file = f'{today_date}_vader_sentiment_analysis.csv'
textblob_file = f'{today_date}_textblob_sentiment_analysis.csv'

# Load both CSV files into DataFrames
vader_df = pd.read_csv(vader_file)
textblob_df = pd.read_csv(textblob_file)

# Merge the two DataFrames on 'Headline' to compare sentiment results for the same headlines
combined_df = pd.merge(vader_df, textblob_df, on="Headline", suffixes=('_vader', '_textblob'))

# Create an index for the x-axis representing the position of the headline from the top of the page
combined_df['Index'] = range(len(combined_df))

# Calculate moving averages for smoother sentiment curves
window_size = 6  # You can adjust this for smoother curves
combined_df['VADER_MA'] = combined_df['Compound'].rolling(window=window_size).mean()
combined_df['TextBlob_MA'] = combined_df['Polarity'].rolling(window=window_size).mean()

# Calculate a longer moving average for the general trend of VADER Compound scores
long_window_size = 20  # Larger window for a smoother trend
combined_df['VADER_Long_MA'] = combined_df['Compound'].rolling(window=long_window_size).mean()

# Set a professional plot style using seaborn
sns.set(style="whitegrid")

# Create a figure for visualizing the comparison
plt.figure(figsize=(12, 8))

# Plot the moving average of VADER and TextBlob sentiment scores for a smoother view (first plot)
plt.subplot(2, 1, 1)
plt.plot(combined_df['Index'], combined_df['VADER_MA'], label='VADER Compound (Moving Average)', color='red', linewidth=2)
plt.plot(combined_df['Index'], combined_df['TextBlob_MA'], label='TextBlob Polarity (Moving Average)', color='blue', linewidth=2, alpha=0.4)
plt.plot(combined_df['Index'], combined_df['VADER_Long_MA'], label='VADER Long-Term Trend (Moving Average)', color='green', linestyle='--', linewidth=2)
plt.title('VADER vs TextBlob (Moving Average of Sentiment)', fontsize=14, fontweight='bold')
plt.xlabel('Headline Position (Index)')
plt.ylabel('Moving Average Sentiment Scores')
plt.legend()
plt.grid(True)

# Plot the raw VADER and TextBlob sentiment scores (second plot)
plt.subplot(2, 1, 2)
plt.plot(combined_df['Index'], combined_df['Compound'], label='VADER Compound Score', color='red', marker='o', alpha=0.6)
plt.plot(combined_df['Index'], combined_df['Polarity'], label='TextBlob Polarity', color='blue', marker='x', alpha=0.6)
plt.title('VADER Compound vs TextBlob Polarity (Raw Scores)')
```
